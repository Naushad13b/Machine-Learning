```python
import pandas as pd
```

```python
data=pd.read_csv("new111.csv")
```

```python
data.head(50)
```

```python
#Mean - missed value first put None to NaN in dataset
import numpy as np
def change1(a):
  data[a] = data[a].replace(np.NaN, data[a].mean())
```

```python
#PM2.5	PM10	NO2	SO2	CO	Ozone	Temp	RH	WS	WD	SR
l=["PM2.5","PM10","NO2","SO2","CO","Ozone","RH","Temp","WS","WD","SR"]
for i in l:
  change1(i)
```

```python
# downloading csv file
cities = pd.DataFrame(data)
cities.to_csv('new222.csv')
```
